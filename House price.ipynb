```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))))
print(((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum())
```

```python
df1 = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR))).sum())
print(df1)
```

```python
# Histograms of all features
df1.hist(figsize=(17,20))
plt.show()
```

```python
df1 = df1.select_dtypes(include=np.number)
print(df1.dtypes)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(df1)
X_scaled = pd.DataFrame(X_scaled, columns=df1.columns)
print(X_scaled)
```

```python
# Printing the dependent and independent features
x = X_scaled.drop('Price', axis=1) # Independent features
y = X_scaled['Price'] # Dependent feature
print('Independent features:', x.columns)
print('Dependent feature:', y.name)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['feature'] = x.columns
vif['VIF'] = [variance_inflation_factor(x, i) for i in range(x.shape[1])]
print(vif)
```

```python
X_reduced = X_scaled.drop(['Resale','WashingMachine','Wifi','Microwave','TV','DiningTable','Sofa','Wardrobe','Stadium'], axis=1)
X_reduced
```

```python
sns.pairplot(X_reduced)
plt.show()
```

```python
corr = X_reduced.corr()
print(corr['Price'])
```

```python
x=pd.DataFrame(X_reduced.drop('Price', axis=1))
y=pd.DataFrame(X_reduced['Price'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split (x, y, test_size=0.2, random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
v=pd.DataFrame(reg.coef_,index=['Co-efficient']).transpose()
w=pd.DataFrame(x.columns,columns=['Attribute'])
```

```python
coeff_df=pd.concat([w,v],axis=1,join='inner')
coeff_df
```

```python
y_pred=reg.predict(x_test)
y_pred=pd.DataFrame(y_pred,columns=['Predicted'])
y_pred
```

```python
y_test
```

```python
from sklearn import metrics
print('Mean absolute error :',metrics.mean_absolute_error(y_test,y_pred))
print('mean sq error:',metrics.mean_squared_error(y_test,y_pred))
print("root mean square error:",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
